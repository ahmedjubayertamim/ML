```python
# Checking Multiple Conditions (and, or)
temperature = float(input("Enter the temperature: "))

if temperature > 30 and temperature < 40:
    print("It's a Fever.")
elif temperature >= 40:
    print("It's extremely Fever!")
else:
    print(" your temperature is normal.")
```

```python
# Nested if Statement
num = int(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
    if num % 2 == 0:
        print("It is also even.")
    else:
        print("It is odd.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

for statement


```python
#Basic for Loop
for i in range(13):
    print("Iteration:", i)
```

```python
# for Loop Over a List
fruits = ["Apple", "Banana", "Cherry","mango","Pineapple"]

for fruit in fruits:
    print(fruit)
```

```python
# Looping Through a String
word = "Bangladesh"

for char in word:
    print(char)
```

```python
# Looping with range(start, stop, step)
for i in range(1, 10, 3):
    print(i)
```

```python
# Looping Over a Dictionary
student = {"name": "Ahmed", "age": 22, "grade": "A"}

for key, value in student.items():
    print(key, ":", value)
```

```python
# Using for Loop with enumerate()
colors = ["Red", "Green", "Blue","Cyan"]

for index, color in enumerate(colors):
    print(index, ":", color)
```

```python
# Nested for Loop
for i in range(1, 2):
    for j in range(1, 4):
        print(f"i={i}, j={j}")
```

```python
# for Loop with break and continue
for num in range(1, 10):
    if num == 5:
        continue  # Skips 5
    if num == 8:
        break  # Stops loop at 8
    print(num)
```

Exercise 1: Write a Python program to find the sum of the first N natural numbers using a for loop.
Example:
Input: N = 5
Output: 15

```python
# Function to calculate the sum of first N natural numbers
def sum_natural_numbers(N):
    total = 0
    for i in range(1, N + 1):
        total += i
    return total


N = int(input("Enter a positive integer: "))

print("Sum of first", N, "natural numbers is:", sum_natural_numbers(N))
```

Exercise 2: Write a Python program that reverses a string using a for loop.
Example:
Input: "Python"
Output: "nohtyP"

```python
# Function to reverse a string using a for loop
def reverse_string(s):
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str
    return reversed_str

s = input("Enter a string: ")

print("Reversed string:", reverse_string(s))
```

Exercise 3: Define a method/function which will take two lists of numbers and returns a list (sum of two lists)

Example:
Input: [3,4,5,1] and [6,7,2,8]
output: [9,11,7,9]

```python
# Function to sum two lists element-wise
def sum_lists(list1, list2):
    return [list1[i] + list2[i] for i in range(min(len(list1), len(list2)))]

list1 = [3, 4, 5, 1]
list2 = [6, 7, 2, 8]

print("Summed List:", sum_lists(list1, list2))
```

Exercise 4: Write a method which will take the full name as input and return both the first name and last name.

Example:
Input: Arafat Rahman Sunny
Output: First name: Arafat, Last name: Sunny

```python
# Function to extract first and last name
def get_first_last_name(full_name):
    parts = full_name.split()  # Split by spaces
    first_name = parts[0]  # First word
    last_name = parts[-1]  # Last word
    return first_name, last_name

full_name = input("Enter full name: ")

first_name, last_name = get_first_last_name(full_name)
print("First name:", first_name)
print("Last name:", last_name)
```
